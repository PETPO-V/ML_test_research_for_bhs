# ЗАДАЧА 1

**Условие:** Дана строка s. Найдите длину самой длинной подстроки без повторяющихся символов. 

```python
def subarray(s):   
    hashset = set()
    
    left = 0
    res = 0
    
    for right in range(len(s)):
        while s[right] in hashset:
            hashset.remove(s[left])
            left += 1
        
        hashset.add(s[right])
        res = max(res, right - left + 1)
    
    return(res)


s = input()
subarray(s)
```

# ОБЪЯСНЕНИЕ АЛГОРИТМА

В данной задаче для оптимального решения мы можем использовать метод "Скользящего окна" (Sliding Window). 
Передам логику своего решения в текстовом виде:
1. Создаём пустой хэшсет, для определения, встречи повторяющегося символа.
2. Инициализируем переменные left (указатель для скользящего окна) и res (для обновления результата)

3. Далее при помощи цикла for идём по строке s переставляя right (указатель для скользящего окна)
4. Пока символ на указателе right находится в hashset (А значит это дубликат), мы удаляем из хэшсета символ на указателе left и смещаем left направо

5. Когда мы выходим из цикла while в хэшсет добавляем символ под указателем right
6. И наконец, проверяем, больше ли текущая подстрока или мы уже находили большую

# ЗАДАЧА 2

**Условие:** Вы поднимаетесь по лестнице. Чтобы добраться до самого верха, вам необходимо пройти n ступеней. Каждый раз, вы можете подняться либо на 1 ступеньку, либо на 2. Как много различных способов подняться на самый верх у вас есть?

```python
def stairs(n):
    if n == 0 or n == 1:
        return n
    
    n1 = 1
    n2 = 2
    
    for i in range(3, n + 1):
        n1, n2 = n2, n1 + n2
    
    return(n2)



n = int(input())
stairs(n)
```
